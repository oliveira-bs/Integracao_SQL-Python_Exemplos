# Aula06 (Manipulaçao de Colunas)

Adicionar coluna na tabela

```python
cursor.execute(f"alter table {table}\
    add column profissao varchar(10);")
```

```python
describe = cursor.execute(f"desc {table}")
desc = pd.DataFrame(cursor.fetchall())
desc.head(10)
```

```python
df = pd.read_sql(f"SELECT * FROM {table}", connection)
df.head()
```

Remover coluna

```python
cursor.execute(f"alter table {table}\
    drop column profissao;")

describe = cursor.execute(f"desc {table}")
desc = pd.DataFrame(cursor.fetchall())
desc.head(10)
```

```python
df = pd.read_sql(f"SELECT * FROM {table}", connection)
df.head()
```

Escolhendo a posição da coluna

```python
cursor.execute(f"alter table {table}\
    add column profissao varchar(10) after nome ;")

describe = cursor.execute(f"desc {table}")
desc = pd.DataFrame(cursor.fetchall())
desc.head(10)
```

```python
df = pd.read_sql(f"SELECT * FROM {table}", connection)
df.head()
```

```python
cursor.execute(f"alter table {table}\
    add column codigo int first;")

describe = cursor.execute(f"desc {table}")
desc = pd.DataFrame(cursor.fetchall())
desc.head(10)
```

Modificando linhas incorretas

```python
cursor.execute(f"""UPDATE {table_1}
SET nome = 'HTML5' 
WHERE idcurso = '1';""")

cursor.execute(f"""UPDATE {table_1}
SET nome = 'PHP', ano = '2015' 
WHERE idcurso = '4';""")

cursor.execute(f"""UPDATE {table_1}
SET nome = 'Java', carga = '40', ano = '2015' 
WHERE idcurso = '5'
limit 1;""") #limit indica que a atualização dos daos só poderá ocorrer para os n primeiros registros/linhas
```

```python
df = pd.read_sql(f"SELECT * FROM {table_1}", connection)
df.head(10)
```

Modificando varios registros/linhas por vez

```python
cursor.execute(f"""UPDATE {table_1}
SET ano = '2050', carga = '800' 
WHERE ano = '2018';""")
```

```python
df = pd.read_sql(f"SELECT * FROM {table_1}", connection)
df.head(10)
```

Modificando um unico registro/linha por vez

```python
cursor.execute(f"""UPDATE {table_1}
SET ano = '2018', carga = '0' 
WHERE ano = '2050'
limit 1;""")
```

```python
df = pd.read_sql(f"SELECT * FROM {table_1}", connection)
df.head(10)
```

Remover uma linha

```python
cursor.execute(f"""DELETE from {table_1} 
WHERE idcurso = '8';""")
```

```python
df = pd.read_sql(f"SELECT * FROM {table_1}", connection)
df.head(10)
```

Remover varios registros/linhas

```python
cursor.execute(f"""DELETE from {table_1} 
WHERE ano = '2050'
limit 2;""")
```
